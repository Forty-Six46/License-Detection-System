```python
from roboflow import Roboflow
from ultralytics import YOLO

# Initialize Roboflow
from roboflow import Roboflow
rf = Roboflow(api_key="twDaQTPidBhprgqrUi0z")
project = rf.workspace("license-plate-detection-hz8kh").project("license-detection-pvfm9")
version = project.version(3)
dataset = version.download("yolov8")

# Load YOLOv8 model (choose size: nano, small, medium, large, xlarge)
model = YOLO("yolov8n.pt")  # e.g., yolov8n.pt, yolov8s.pt, etc.

# Train the model
results = model.train(
    data=f"{dataset.location}/data.yaml",
    epochs=30,
    imgsz=640,
    batch=16,
    name="yolov8_custom"
)
```

```python
!pip install easyocr
from google.colab import files
import easyocr
import cv2
from IPython.display import Image, display

# Upload image
uploaded = files.upload()
image_path = next(iter(uploaded))

# Run EasyOCR
reader = easyocr.Reader(['en'], gpu=True)
results = reader.readtext(image_path)

# Print results
for (bbox, text, prob) in results:
    print(f"Text: {text}, Confidence: {prob:.2f}")

# Display image with boxes
image = cv2.imread(image_path)
for (bbox, text, prob) in results:
    (tl, tr, br, bl) = bbox
    tl = (int(tl[0]), int(tl[1]))
    br = (int(br[0]), int(br[1]))
    cv2.rectangle(image, tl, br, (0, 255, 0), 2)
    cv2.putText(image, text, (tl[0], tl[1] - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)
cv2.imwrite('output.jpg', image)
display(Image('output.jpg'))
```
